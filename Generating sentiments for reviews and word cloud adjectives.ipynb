```python
import re
import nltk
import pandas as pd
```

```python
df = pd.read_csv('C:/A-backup/NMIMs/Artificial Intelligence/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')
```

```python
df.head()
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
```

```python
adjectives = []   #list for adjectives

placeholder = [] #temporary list for holding values

sentences = []  #list which contains review wise adjectives

for i in range(len(df['reviews.text'])):
    
    for word,pos in nltk.pos_tag(word_tokenize(df['reviews.text'][i])):
        
        if pos=="JJ" or pos=="JJR" or pos=="JJS":
            
            placeholder.append(lemm.lemmatize(word))
            
            adjectives.append(lemm.lemmatize(word))
            
            Text = ' '.join(placeholder)
    
    placeholder = []        
    
    sentences.append(Text)
            
```

```python
adjectives
```

```python
#creating a series out of adjectives frequency

data = pd.Series(adjectives).value_counts()

data
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#Generate word cloud
wc = WordCloud(width=400, height=330, max_words=100, background_color='black').generate_from_frequencies(data)

plt.figure(figsize=(14, 10))

plt.imshow(wc, interpolation='bilinear')

plt.axis('off')

plt.show()
```

```python
from textblob import TextBlob
```

```python
Amazon = pd.DataFrame(sentences,columns = ['Adjectives'])
```

```python
#creating a list of sentiments based on textblob score

Sentiment =[]

for i in range(len(Amazon)):
    
    score = TextBlob(Amazon['Adjectives'][i]).sentiment.polarity
    
    if score > 0:
        
        Sentiment.append("Positive")
    
    else:
        
        Sentiment.append("Negative")
    
```

```python
#Reorganizing the Amazon dataset 

Amazon['Sentiments'] = Sentiment

Amazon['Reviews'] = df['reviews.text']

cols = ['Reviews','Adjectives','Sentiments']

Amazon = Amazon[cols]
```
